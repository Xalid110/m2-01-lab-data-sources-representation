```python
def validate_registrations(entries):
    valid_entries = []
    invalid_entries = []
    
    for entry in entries:
        try:
            user = isinstance(entry.get("username"), str) and entry["username"].strip() != ""
            email = isinstance(entry.get("email"), str) and "@" in entry["email"]
            
            age = int(entry.get("age"))
            age_valid = age > 0
            
            if user and email and age_valid:
                valid_entries.append(entry)
            else:
                invalid_entries.append(entry)
                
        except (ValueError, TypeError):
            invalid_entries.append(entry)
    
    return valid_entries, invalid_entries

valid_entries, invalid_entries = validate_registrations(registration)
print("Valid:", len(valid_entries))
print("Invalid:", len(invalid_entries))
```

```python
import numpy as np
import pandas as pd

rows = 200_000
cols = 20

data = np.random.rand(rows, cols)
columns = [f"feature_{i}" for i in range(cols)]
df = pd.DataFrame(data, columns=columns)

df.head()
```

```python
#operation a
%time 
df["feature_0"].mean()
```

```python
#operation b
%time
total = 0
for i in range(10_000):
    total += df.iloc[i]["feature_0"]
```

```python
arr=df.to_numpy()
```

```python
#operation c
%time
arr[:, 0].mean()
```

```python
#operation d
%time
total = 0
for i in range(10_000):
    total += arr[i, 0]
```

1.Because pandas operates on the entire column at once and vectorized,but in iteration each time it creates series,index alignment and data processed each time.
2.Because NumPy is also vectorized and runs in base of C code.Secondly it works without creating series.Additionally it doesn't store row/column labels and not perform index alignment.
3.NumPy stores data in a block of homogeneous memory (same dtype).Pandas is built on top of NumPy but adds index labels,column labels,type handling,metadata management.
Row-major storage stores data row by row in memory. All values of the first row are stored first, followed by the second row, and so on. It is efficient when entire rows are accessed together.

Column-major storage stores data column by column in memory. All values of the first column are stored first, followed by the second column, and so on. It is efficient when entire columns are accessed together.

In short, row-major prioritizes row access, while column-major prioritizes column access.

```python
import pandas as pd
data = [
    ["The Silent Patient", "Alex Michaelides", "Paperback", "Celadon Books", "US", 15.99],
    ["The Silent Patient", "Alex Michaelides", "E-book", "Celadon Books", "US", 9.99],
    ["Where the Crawdads Sing", "Delia Owens", "Paperback", "G.P. Putnam's Sons", "US", 17.50],
    ["Where the Crawdads Sing", "Delia Owens", "E-book", "G.P. Putnam's Sons", "US", 11.99],
    ["Educated", "Tara Westover", "Paperback", "Random House", "US", 16.00],
    ["Becoming", "Michelle Obama", "Paperback", "Crown Publishing Group", "US", 18.00],
    ["Becoming", "Michelle Obama", "E-book", "Crown Publishing Group", "US", 12.99],
    ["Normal People", "Sally Rooney", "Paperback", "Faber & Faber", "UK", 14.50],
    ["Normal People", "Sally Rooney", "E-book", "Faber & Faber", "UK", 10.99],
    ["Circe", "Madeline Miller", "Paperback", "Little, Brown & Co.", "US", 16.50],
    ["Circe", "Madeline Miller", "E-book", "Little, Brown & Co.", "US", 10.50],
    ["The Testaments", "Margaret Atwood", "Paperback", "Nan A. Talese", "US", 19.00]
]

columns = ["title", "author", "format", "publisher_name", "publisher_country", "price"]
flat_df = pd.DataFrame(data, columns=columns)
flat_df
```

```python
publishers = flat_df[["publisher_name", "publisher_country"]].drop_duplicates().reset_index(drop=True)
publishers["publisher_id"] = range(1, len(publishers)+1)

books = flat_df.merge(publishers, on=["publisher_name", "publisher_country"])
books = books[["title", "author", "format", "publisher_id", "price"]]
```

```python
reconstructed_df = books.merge(publishers, on="publisher_id")[
    ["title", "author", "format", "publisher_name", "publisher_country", "price"]
]
is_identical = reconstructed_df.equals(flat_df)#for checking merged with original
is_identical
```

```python
books_docs = []

for _, book in flat_df.iterrows():
    books_docs.append({
        "title": book["title"],
        "author": book["author"],
        "publisher": {
            "name": book["publisher_name"],
            "country": book["publisher_country"]
        },
        "format": book["format"],
        "price": book["price"]
    })

books_docs[:3]
```

```python
#step4
#1
author_name = "Alex Michaelides"

# Flat
flat_author = flat_df[flat_df['author'] == author_name]
print(flat_author[['title','format','price']])

# Normalized
normalized_author = books[books['author'] == author_name].merge(publishers, on='publisher_id')
print(normalized_author[['title','format','price','publisher_name']])

# Document-oriented
doc_author = [b for b in books_docs if b['author'] == author_name]
print(doc_author)
```

```python
#2
# Flat
avg_price_flat = flat_df['price'].mean()
print("Flat average price:", avg_price_flat)

# Normalized
avg_price_normalized = books['price'].mean()
print("Normalized average price:", avg_price_normalized)

# Document-oriented
all_prices = [b['price'] for b in books_docs] 
avg_price_docs = sum(all_prices) / len(all_prices)
print("Document average price:", avg_price_docs)
```

```python
#3
# Flat
flat_df.loc[flat_df['publisher_name']=="Celadon Books",'publisher_country'] = "Canada"
print(len(flat_df[flat_df['publisher_name'] == "Celadon Books"]))

# Normalized: update publisher country in publishers_df
publishers.loc[publishers['publisher_name']=="Celadon Books",'publisher_country'] = "Canada"
# Only 1 row in publishers_df changed; joined data automatically reflects this
print(len(publishers[publishers['publisher_name'] == "Celadon Books"]))

# Document-oriented
count_updates = 0
for book in books_docs:
    if book['publisher']['name'] == "Celadon Books":
        book['publisher']['country'] = "Canada"
        count_updates += 1
print("Documents updated:", count_updates)
```

1.For finding specific author,flat was easiest.For average price flat and normalized are equally easiest.In publisher change again flat and normalized are equally easiest.
2.I would choose normalized beacuse only one row per publisher needs updating,so makes it easier and faster.
3.Document-oriented because it is self-contained we don't need joins
